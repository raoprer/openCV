```python
import cv2
image = cv2.imread("bird.jpg", cv2.IMREAD_COLOR)
cv2.imshow("image", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image.shape
```

```python
RGB_img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(RGB_img)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
gray.shape
```

Extracting B, G, R channels from an image

```python
B, G, R = cv2.split(image) #splits the image into B, G, R arrays stacked behind one other (Color channel)
```

```python
cv2.imshow("Blue", B)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow("Green", G)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imshow("Red", R)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
added = cv2.add(B, G)
cv2.imshow("Added", added)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
weighted = cv2.addWeighted(B, 0.5, G, 0.5, 0.2)
cv2.imshow("Added", weighted)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
